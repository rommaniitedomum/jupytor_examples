## 맷플롯 라이브러리 ##

- 데이터 시각화를 위한 라이버르러
- 라이버르러에서 세세하게 지원하는 기능이 많지만 주로 사용하는 기능은 한정해서 임포트

```python
import matplotlib.pyplot as plt

plt.plot([1,2], [2,4])
plt.show()  # 그래프 그리기 
```

```python
# 하나의 축만 숫자 리스트로 입력하면 대응 값은 디펄트 인덱스 0 부터 대응 갯수로 자동 저장 
y = [1,2,3,4]
plt.plot(y)
plt.show()
```

```python
#  plt 함수의 label 매게 변수에 지정된 문자열은 범례
plt.plot([1,2,3,4], [1,4,9,16], label="Square")
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.legend() # 라벨(범례) 표시 
plt.show()
```

```python
import numpy as np
#  linespace() 함수를 사용하여 넘파이 배열 생성 
x = np.linspace(0, 10, 100)

y1 = x
y2 = 2 * x
y3 = 0.5 * x
y4 = 3 * x

plt.plot(x, y1, label= 'y = x')
plt.plot(x, y2, label= 'y = 2x')
plt.plot(x, y3, label= 'y = 0.5x')
plt.plot(x, y4, label= 'y = 3x')
plt.legend(loc="upper center", ncol=2)

plt.title('Simple Linear Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.plot([1,2,3,4],[3,6,9,12])
plt.xlim([0,15])
plt.ylim([0,15])
plt.show()

#  plt 축 0.0.0.
```

## 축의 스케일 ##


- 선형 스케일: 가장 많이 사용되는 스케일로, 축에 일정 간격으로 눈금을 표시
- 로그 스케일: 축의 간격을 로그 스케일로 변환
    1. 값의 크기 차이가 큰 경우 사용
    2. 예를 들어 지수적으로 증가하는 데이터나 큰 범위의 데이터를 다룰 때 사용
- 로그-선형 스케일: x축을 로그 스케일, y축을 선형 스케일로 표시
    1. 주로 x축이 시간 또는 크기인 경우 사용
    2. 예를 들어 시간에 따른 데이터나 크기에 따른 데이터를 나타낼 때 적합
- 선형-로그 스케일: x축을 선형 스케일, y축을 로그 스케일로 표현
    1. 주로 y축의 값 범위가 큰 경우 사용
    2. 예를 들어 대부분의 값은 작은 범위에 있지만 일부 값이 매우 큰 경우 사용

```python
# 1차원 배열에서 지수 함수 계산 
# 자연로그의 밑 e는 수학에서 약 2.71828로 나타나는 무리수로, 지수 함수와 로그 함수의 중요한 기반이 되는 상수.
# 이에 따라 e^2 자연로그의 밑 e를 2번 곱한 값. 이를 계산하면: 2.71828^2 = 7.38906
# e^2은 시간이 2단위 지났을 때, 어떤 초기값이 2배로 증가했을 때의 결과를 나타낸다.
# 만약 100의 인구가 연속적으로 연 100%씩 증가한다면, 2년 후 인구는 100×𝑒^2 (약 738.9명)가 됩니다.

array_2 = 2
array_3 = 3
exp_array_1 = np.exp(array_2)
exp_array_2 = np.exp(array_3)

print("배열과 지수 함수 결과", exp_array_1)
print("배열과 지수 함수 결과", exp_array_2)
```

```python
# plt 서브플롯 하나의 그ㅐ프창 피겨  안에 여러 그래프 배치 
# 파라미터 엔로우 엔콜 엔인덱스 엔 인덱 2개 2행으로 나눔 
# 몇 칼럼 나눌지 지정
# 나뉜 격자중에서 몇번쨰 위치에 그래프 그릴지 지정 

x = np.array([0,1,2,3])
y = np.array([3,8,1,10])


plt.subplot(2, 1, 1)
plt.plot(x,y)


# nrow ncol nindex

```

```python
x = np.linspace(1, 10, 100)
y = np.exp(x)

#  linear scale
plt.subplot(2, 2, 1) # 2col 1 영역 그래프
plt.plot(x,y)
plt.title('sun hyung scale') 

# log scale 
plt.subplot(2,2,2)
plt.plot(x,y)
plt.yscale('log') # y axis 
plt.title('logarithmic scale -yaxis') 

# log  linearscale 
plt.subplot(2,2,3)
plt.plot(x,y)
plt.xscale('log') # y axis 
plt.title('log-linear scale -xaxis') 

# linear log scale 
plt.subplot(2,2,4)
plt.plot(x,y)
plt.yscale('log') # y axis 
plt.xscale('linear') # y axis 
plt.title('linear logarithmic scale -xaxis') 

plt.tight_layout() # sub graph layout for plotting
plt.show()
```

```python
# 선 스타일 지정 
# plot() 함수의 세번쨰 인수 값으로 간략화한 문자열을 입력하여 선 종류 설정 
plt.plot([1,2,3],[4,4,4], '-', label='solid')
plt.plot([1,2,3],[3,3,3], '--', label='dashed')

# plot 함수의 lin style 값을 문자열로 입력하여 선 종류설정 
plt.plot([1,2,3],[2,2,2], linestyle='dotted', label='solid')
plt.plot([1,2,3],[1,1,1], linestyle='dashdot', label='dashed')


# 축 범위와 범례 위치 설정 
plt.xlim([0.8, 3.2])
plt.ylim([0.5, 5.0])
plt.legend(loc='upper center', ncol=4)
plt.show()
```

```python
# 마커 설정 
plt.plot([4,5,6], 'b') # blue
plt.plot([3,4,5], 'ro', linestyle='-') # red round
plt.plot([2,3,4], marker='s') # square
plt.plot([1,2,3], marker='D') # diamond
plt.plot([0,1,2], marker='^') # triangle
plt.plot([5,6,7], marker='$A$') # $char$

plt.show()
```

```python
# color plot color param values 
plt.plot([1,2,3,4], [2.0,3.3,6.3,10.5], color='indigo') # 컬러 이름 지정 
plt.plot([1,2,3,4], [2.0,3.1,5.3,8.5], color='g') # 컬러 이름 이니셜  
plt.plot([1,2,3,4], [2.0,2.5, 3.3,4.5], color='#FF0000') # 16bit컬러 이름 이니셜  
```

```python
# title option : fontsize, color , fontweight background ~~~~
# loc : 제목의 위치 - 기본값을 center 이고 , left , right 등으로 변경 가능 
# pad 제목과 그래프의 간격 

plt.plot([1,2,3,4],[1,4,9,16])
plt.title("Graph Title", fontsize=16, color='blue', fontweight='bold', loc='center', pad=100, backgroundcolor='lightgray')
plt.show()
```

```python
# super title
x=[1,2,3,4,5]
y1= [2,4,6,8,10]
y2=[1,2,1,2,1]

# 첫번째 서브 그래프 
plt.subplot(2,1,1)
plt.plot(x, y1)
plt.title('Graph sub title 1', loc='left')

# second sub graph
plt.subplot(2,1,2)
plt.plot(x, y2)
plt.title('Graph sub title 2', loc='right')

plt.suptitle('graph')

plt.tight_layout()
plt.show()
```

```python
# 막대 그래프 

x = [1,2,3] # x 축 갯수 
years=['2030','2040','2050']
values=[300,100,700]

# 막대 그래프 색상 설정 
plt.bar(x, values, color=['r', 'g', 'b'])

# 눈금표시 : xticks()
plt.xticks(x, years)
plt.show()
```

```python
# 격자 grid 지정 : grid()
#params== linestyle linewidth color alpha , axis

plt.plot([1,2,3,4], [10,5,10,5])
plt.grid(linestyle='--', linewidth=0.5, color='gray', alpha=0.5, axis='x') 
plt.show()
```

```python
# 수치 텍스트 추가  text()
category=['A','B', 'C','D']
values=[25,40,30,20]

# 막대 그래프 생성 
plt.bar(category, values, color='lightblue')

# plt.text(0, values[0] + 1, str(values[0]), ha='center', fontsize=12)
# ha parameter: 수평 , va 파라미터 : 수직 
for i, value in enumerate(values):
     plt.text(i, values[i] + 1, str(values[i]), ha='center', va='bottom', fontsize=12)


plt.xlabel('category')
plt.ylim([0, 50])
plt.ylabel('values')
plt.show()
```
