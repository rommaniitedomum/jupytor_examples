## 범주형 산점도 그래프 ##
- 시본의 대표적인 산점도 그래프는 dcatterplot 함수다.
- 연속형 변수와 변수의 관계를 나타내는 그래프다.


## 연속형 변수와 범주형 변수 ##
- 연속형 변수: 온도와 키, 그리고 시간 등을 연속형 변수라 한다. 연속적인 값을 가지며 무한이 많은 값이 있다. 실수로 표현하며 범위 내에서 어떠한 값이라도 취할 수 있다. 일반적으로 양적인 정보를 제공하며 측정 가능한 크기나 양을 가지고 있다.


- 범주형 변수: 명목형과 순서형으로 나눌 수 있다.
 - 명목형 변수는 서로 구별되는 범주를 가진다. 예를 들어 성별, 혈액형 등이 있다.
 - 순서형 변수는 범주 간에 순서 또는 등급이 있다. 예를 들어 학력 수준, 만족도 등이 있다.


- 범주형 산점도 그래프는 stripplot, swarmplot이 있다.
 - stripplot: 주어진 범주형 변수의 분포를 점으로 나타내는데 사용한다. 주로 하나의 변수에 대한 데이터를 표시하며, 점이 겹칠 수 있다. 데이터의 밀도를 표현한다.
 - swarmplot: 데이터 포인트가 겹치지 않도록 적절한 위치에 배치하여 시각화 한다. 가로축 상의 각 카테고리에 대해 데이터 포인터가 서로 중첩되지 않게 나열한다. 이를 통해 데이터 포인트를 명확하게 볼 수 있고, 데이터 분포를 시각적으로 파악할 수 있다.

```python

tips = sns.load_dataset('tips')

# figure에 2개의 서브 플롯 생성
fig = plt.figure(figsize=(15, 5))
ax1 = fig.add_subplot(1, 2, 1) # 로우가 한개이고 컬럼이 두개이며 첫번째 그래프 표시
ax2 = fig.add_subplot(1, 2, 2)

# stripplot 생성
sns.stripplot(x='day', y='tip', hue='sex', data=tips, alpha=0.7, ax=ax1)

# swarmplot 생성
sns.swarmplot(x='day', y='tip', hue='sex', data=tips, alpha=0.7, ax=ax2, palette='Set2')

ax1.set_title('Strip Plot of Tip by Day and Gender')
ax2.set_title('Swarm Plot of Tip by Day and Gender')

plt.show()


# hue : hue 로 지정된 데이터의 카테고리별로 색상을 다르게 하는 포인트 구분 
# data 시각화할 데이터를 포함하는 데이터 프레임을 지정 
# order : x 축에 나타난 범주의 순서를 지정, 범주의 이름의 리스트를 전달하여 순서를 정함 
# hue_order: hue 에 따른 범주의 순서를 지정 
# dodge True 로 설정하면 hue 에 따른 범주들이 나란히 그려짐 
# orient: 그래프 방향을 설정. x 또는 y 로 설정하여 데이터 포인트 수평 혹은 수직으로 그려지게함 
```

```python
# 빈도그래프: countplot() - 막대 그래프
# 범주형 변수의 카테고리별 빈도를 막대그래프로 표현

# figure 2개의 서브 플롯 

fig = plt.figure(figsize=(15, 5))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

# 식사 이뤄진 시간대 파악 
# 파라미터 x 축 변수 , 데이터셋 axe 객체 
sns.countplot(x='time', data=tips, ax=ax1)

# 식사가 이뤄진 시간대 파악과 식사가 이뤄진 요일로 색상별 막대 그래프 표현 
# x 축 변수, hue 색상 분류, 데이터셋, 색상설정, exe 객체 

sns.countplot(x='time', hue='day', data=tips, palette='Set2', ax=ax2)

ax1.set_title('Count Plot of Meal Time')
ax2.set_title('Count Plot of Meal Time by Day')

plt.show()
```

## 선형 회귀선이 있는 산점도 그래프 ##
- 두 변수 간의 선형 관계를 시각화하는 데 사용
- 산점도 그래프와 함께 선형 회귀선을 함께 그려주므로 두 변수 간의 추세를 확인하기 용이


## 주요 매개변수 ##
- x_bins: x 데이터를 지정된 개수의 구간으로 나누어, 각 구간의 중앙값을 기준으로 y값을 추정
- order: 회귀선의 다항식 차수를 지정. 기본값은 1이며, order=2로 설정하면 2차 곡선 회귀선을 표시
- scatter: False로 설정하면 산점도를 그리지 않고 회귀선만 표시
- fit_reg: False로 설정하면 회귀선을 그리지 않고, 산점도만 표시. 최근 시본 버전에서는 scatter_kws와 line_kws로 대체되고 있음
- line_kws: 회귀선의 스타일을 지정하는 데 사용. 예를 들어 선의 색상, 두께 등을 설정
- scatter_kws: 산점도의 점 스타일을 지정하는 데 사용. 점의 크기, 색상 등을 설정
- dropna: True로 설정하면 x와 y에 NaN이 있는 행을 제외


```python
# 한명당 지불한 총 식사 비용과 식사에 대해 받은 팁의 상관관계

# figure에 2개의 서브 플롯 생성 
fig = plt.figure(figsize=(15,5))

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.regplot(x= 'total_bill', y= 'tip', data= tips, color='blue', scatter_kws={'s': 50, 'alpha':0.5}, line_kws={'linestyle':'--'}, ax=ax1 )

sns.regplot(x= 'total_bill', y= 'tip', data= tips, color='blue', scatter_kws={'s': 50, 'alpha':0.5}, line_kws={'linestyle':'--'}, ax=ax2, fit_reg=False )

fig.suptitle('scatter plots with regs', fontsize=16)
ax1.set_title('rgs')

ax2.set_title('no rgs')

plt.show()
```

## 히스토그램과 커널 밀도 추정 그래프 ##


- 히스토그램: 연속형 데이터를 일정한 간격으로 나누어 각 구간에 속하는 데이터 빈도를 막대로 표현한 그래프
- 각 막대는 데이터 분포를 나타내며 높이는 해당 구간의 데이터 개수를 표현


- 커널 밀도 추정 그래프: 데이터 분포를 나타낸다는 점에서 히스토그램과 비슷함. 하지만 막대 대신 부드러운 곡선을 사용
- 커널 함수는 각 데이터 포인트 주위에 작은 확률 분포를 만들고, 이를 모두 합혀여 전체 데이터의 분포를 표현


## 주요 파라미터 ##


- bins: 막대의 개수를 지정. 기본값은 데이터에 따라 자동으로 설정되지만, 직접 숫자를 지정하여 막대의 개수를 조절할 수 있음.
- binwidth: 히스토그램에서 각 막대의 너비를 지정. 이 값을 지정하면 bins 매개변수 대신 각 막대의 너비를 기준으로 히스토그램이 표현됨
- hue: 데이터의 카테고리별로 히스토그램을 색상으로 구분하여 표현. 범주형 변수를 지정하면, 각 범주에 대해 별도의 히스토그램이 색상으로 구분되어 나타남
- multiple: 여러 히스토그램이 그려질 때 이들을 어떻게 처리할지 지정
 - layer: 겹쳐 그리기
 - dodge: 옆으로 나란히 그리기
 - stack: 쌓아 그리기
 - fill: 히스토그램이 합쳐저 전체적으로 1이 되도록 크기를 비율로 조정
- kde: True로 설정하면 컨널 밀도 추정 곡성을 히스토그램 위에 추가로 표현. 이 곡선은 히스토그램보다 데이터의 분포를 더 부드럽게 표현
- stat: 히스토그램의 막대 높이를 어덯게 계산할지를 결정
 - count: 각 빈에 속하는 데이터 포인트의 개숙를 표시. 이 설정은 기본값으로, 막대의 높이가 해당 범위에 속하는 데이터 포인트의 수를 표현
 - frequency: 각 빈의 데이터 포인트 개수를 빈 너비로 나눈 값을 표시. 즉, 빈이 넓을수록 빈의 빈도가 더 낮게 표시.
 - probability or proportion: 히스토그램의 막대 높이가 전체 데이터 포인트 수에 대해 해당 빈의 비율을 나타내도록 정규화.
 - percent: 히스토그램의 막대 높이가 전체 데이터 포인트 수에 대해 해당 빈의 비율을 백분율로 표시.
 - density: 히스토그램의 막대 높이를 전체 면적이 1이 되도록 정규화. 이 설정은 데이터의 밀도를 시각화하는 데 사용되며, 커널 밀도 추정 그래프와 유사한 해석을 제공


- element: 히스토그램의 스타일 표현
 - bars: 막대
 - step: 계단형
 - poly: 폴리곤 형태


- fill: True로 설정하면 막대를 채운 형태로, False로 설정하면 테두리만 표현
- shrink: 막대의 너비를 조정. 예를 들어 shrink=0.8로 설정하면 막대의 너비가 80%
- common_norm: True로 설정하면 누적 히스토그램을 생성. 각 막대는 그 이전까지의 합계를 나타내므로, 데이터의 누적 분포를 볼 수 있음.
- cumulative: True로 설정하면 누적 히스토그램을 생성. 각 막대는 그 이전까지의 합계를 나타내므로, 데이터의 누적 분포를 볼 수 있음.

```python
# 히스토 그램과 커널 밀도 추정 그래프 통합 
# histplot 

sns.histplot(tips['tip'], bins=30, color='skyblue', kde=True)
plt.title('histogram')
plt.show()
```
